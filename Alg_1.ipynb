### Задание 1

Реализуйте программу, выдающую число ненулевых битов в бинарном представлении числа. Запрещается переводить число в строку, т.е решить задачу необходимо при помощи битовых операций.

```python
x = int(input())
k = 0
while x > 0:
    i = x % 2
    x = x // 2
    if i == 1:
        k += 1
print(k)
```

### Задание 2

Реализуйте программу, выводящую n-й бит в представлении числа. Программа должна работать за O(1).  Запрещается переводить число в строку, т.е решить задачу необходимо при помощи битовых операций.

```python
x, y = input().split()
print((int(x) >> int(y) - 1) & 1)
```

### Задание 3

Реализуйте программу, присваивающую n-му биту в представлении числа значение 1 и выводящую это число. Программа должна работать за O(1).  Запрещается переводить число в строку, т.е решить задачу необходимо при помощи битовых операций.

```python
x, y = input().split()
print(int(x) | 1 << (int(y) - 1))
```

### Задание 4


Реализуйте программу, меняющую значение n-го биту в представлении числа на противоположное и выводящую это число. Программа должна работать за O(1). Нельзя использовать if. Запрещается переводить число в строку, т.е решить задачу необходимо при помощи битовых операций.

```python
x, y = input().split()
print(int(x) ^ (1 << (int(y) - 1)))
```

### Задание 5

Реализуйте программу, которая принимает на вход два числа и выводит число битов, которые надо поменять в представлении первого числа, чтобы получить второе

```python
x, y = input().split()
z = int(x) ^ int(y)
k = 0
while z > 0:
    i = z % 2
    z = z // 2
    if i == 1:
        k += 1
print(k)
```

### Задание 6

(Для последующих двух задач использование битовых операций не принципиально)

Программа  принимает на вход число N и, следующей строкой, массив с числами от 1 до N в случайном порядке. Одно из чисел пропущено, программа должна вывести это число. Программа должна работать за O(N) и использовать O(1) дополнительной памяти



```python
x = int(input())
y = input().split()
full = x
notfull = 0
for i in range(len(y)):
    full += i + 1
    notfull += int(y[i])
print(full - notfull)
```

### Задание 7

Программа  принимает на вход число N и, следующей строкой, массив с числами от 1 до N в случайном порядке. Два числа пропущены, программа должна вывести эти числа в порядке возрастания. Программа должна работать за O(N) и использовать O(1) дополнительной памяти


```python
# Идея: http://espressocode.top/find-two-missing-numbers-set-1-an-interesting-linear-time-solution/
x = int(input())
y = list(map(int, input().split()))
avg = (x * (x + 1) // 2 - sum(y)) // 2
sum_less = 0
sum_more = 0
for i in y:
    if i <= avg:
       sum_less += i
    else:
       sum_more += i
less = ((avg * (avg + 1)) // 2) - sum_less
more = x * (x + 1) // 2 - sum(y) - less
print(less, more)
```

### Задание 8 (доп задание)

Реализуйте программу, принимающую на вход массив с числами, лишь одно из которых встречается нечетное число раз. При помощи битовых операций определите и выведите это число. Запрещается использовать множества, словари и похожие коллекции. Алгоритм должен работать за O(N), где N - длина массива

```python
x = input().split()
y = 0
for i in x:
    y = y ^ int(i)
print(y)
```
